```python
# Import Dependencies
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get SD of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
# Create the plot
plt.plot(x_axis, y_axis, marker="D",color = "green",linewidth = 2, label = "Boston")
#crate labels for the x and y axces
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add a legend
plt.legend()
# Add Error bars
plt.errorbar(x_axis,y_axis,yerr=stdev, capsize = 3)
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis, label='Boston', marker="D", color = "green", linewidth = 2)
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month (OBJECT)")
ax.legend()
ax.grid()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize = 3)
```

```python
# Create the veritcal bar graph
plt.bar(x_axis, y_axis, color = "green", label = 'Boston', yerr = stdev, capsize=3)
# Create Labels
plt.xlabel("Dates")
plt.ylabel("Fare($)")
# Create Title
plt.title("PyBer Fare by Month")
# Add Legend
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color = "green", yerr=stdev, capsize = 3)
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer by Month")
```
