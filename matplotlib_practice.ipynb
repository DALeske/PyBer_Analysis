```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Create values for the X-axis = months of the year
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis values in a list: total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Alternate
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create plot wtih ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot (AND SHOW)
plt.plot(x_axis, y_axis)
plt.show
```

```python
# Create the plat and add a labelf or the legend
plt.plot(x_axis, y_axis, label = 'Boston')
#craete labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set teh y liomit between 0 and 45
plt.ylim(0,45)
# Create a title
plt.title("PyBer Fare by Month")
#Add the Legend
plt.legend()
```

```python
# Create the plot
plt.plot(x_axis, y_axis, marker="D",color = "green",linewidth = 2, label = "Boston")
#crate labels for the x and y axces
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45
plt.ylim(0,45)
# Create a title
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add a legend
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis, label='Boston', marker="D", color = "green", linewidth = 2)
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month (OBJECT)")
ax.legend()
ax.grid()
```

BAR GRAPHS

```python
# Create the veritcal bar graph
plt.bar(x_axis, y_axis, color = "green", label = 'Boston')
# Create Labels
plt.xlabel("Dates")
plt.ylabel("Fare($)")
# Create Title
plt.title("PyBer Fare by Month")
# Add Legend
plt.legend()
```

```python
# Create horizontal bar graph 
plt.barh(x_axis, y_axis, color = "magenta", label = "Boston")
plt.gca().invert_yaxis()
plt.legend()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color = "red", label = "Boston")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("Pyber Fare by Month (OBJECT)")
ax.legend()
```

```python
# Create a horizontal bar graph with OBJECT oriented
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label = "Chicago")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
plt.gca().invert_yaxis()
ax.legend()
```

```python
# Scatter plot
plt.plot(x_axis, y_axis,"o")
```

```python
plt.scatter(y_axis, x_axis, c= "red", label = "Chicago")
plt.legend()
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("Fare by Month")
plt.gca().invert_yaxis()
plt.xlim(0, 50)
```

```python
# Bubble PLot
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# List Comprehension
plt.scatter(x_axis, y_axis, s=[junk*3 for junk in y_axis],c="green")
```

```python
# Bubble plot with Object Oriented

fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[var*5 for var in y_axis], c="skyblue", alpha= 0.8, edgecolors= "black", linewidths = 2, label = "Boston")
ax.legend()
ax.set_title("Ride volume vs Month")
ax.set_xlim(0,50)
plt.gca().invert_yaxis()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
```

```python
# Pie Chart with MATLAB Approach
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8,8))
plt.pie(y_axis,labels=x_axis, explode = explode_values, autopct='%.1f%%', colors = colors)
plt.show()
```

```python
# Pie Chart with object oriented
fig, ax = plt.subplots(figsize=(8,8))
explode_nums = [0,0,0.2,0,0,0,0.2,0,0,0,0,0]
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

ax.pie(y_axis, labels = x_axis, autopct= '%.1f%%',explode=explode_nums, shadow = True, colors = colors)
plt.gca().invert_yaxis()
plt.show()
```
